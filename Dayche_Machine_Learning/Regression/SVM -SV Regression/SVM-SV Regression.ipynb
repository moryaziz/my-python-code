## Initializaion


```python
import numpy as np
import math
from sklearn.svm import SVR
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

#### Create random input and output data

```python
x_train = np.linspace(-math.pi, math.pi, 2000).reshape(-1, 1)
# use reshape(-1,1) to reshape export of linspace to 2D like vector. (it was 1D)
y_train = np.sin(x_train).reshape(-1, 1)
x_test = np.linspace(-math.pi, math.pi, 200).reshape(-1, 1)
y_test = np.sin(x_test).reshape(-1, 1)
plt.plot(x_train, y_train, 'b')
```

#### Normalization

```python
scaler_x = StandardScaler().fit(x_train)
scaler_y = StandardScaler().fit(y_train)
x_train, x_test = scaler_x.transform(x_train), scaler_x.transform(x_test)
y_train, y_test = scaler_y.transform(y_train), scaler_y.transform(y_test)
```

#### Linear regression model

```python
linear_model = SVR(kernel = 'linear')
linear_model.fit(x_train, y_train)

```

```python
y_pre_te = linear_model.predict(x_test)
print(linear_model.support_)
print(linear_model.n_support_)
print(linear_model.coef_, linear_model.intercept_)
plt.plot(x_test, y_pre_te, 'b')
plt.plot(x_test, y_test, 'r')
plt.legend(['prediction', 'actual value'])
```

#### Polynomial kernel

```python
poly_model = SVR(kernel = 'poly') #C = 1, epsilon = 0.1,   degree = 3)
poly_model.fit(x_train, y_train)
y_pre_te = poly_model.predict(x_test)
print(poly_model.support_)
print(poly_model.n_support_)
print(poly_model.dual_coef_[0], poly_model.intercept_)
plt.plot(x_test, y_pre_te, 'b')
plt.plot(x_test, y_test, 'r')
plt.legend(['prediction', 'actual value'])
```

```python
rbf_model = SVR(kernel = 'rbf')
rbf_model.fit(x_train, y_train)
y_pre_te = rbf_model.predict(x_test)
plt.plot(x_test, y_pre_te, 'b')
plt.plot(x_test, y_test, 'r')
plt.legend(['prediction', 'actual value'])
```
